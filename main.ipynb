```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Read Data
data = pd.read_csv('cost_revenue_clean.csv')
```

```python
#Data Description
data.describe()
```

```python
X = pd.DataFrame(data, columns=['production_budget_usd'])
y = pd.DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
#Scatter Plot
plt.figure(figsize=(10, 8))
plt.scatter(X, y, alpha=0.3)
plt.xlim(0, 4.5e8)
plt.ylim(0, 3e9)
plt.xlabel('production_budget_usd')
plt.ylabel('worldwide_gross_usd')
plt.title('Worldwide Gross v/s Production Budget')
plt.show()
```

```python
#Linear Regression
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
#Intercept
theta_0 = model.intercept_[0]
print(theta_0)
```

```python
#Slope
theta_1 = model.coef_[0][0]
print(theta_1)
```

```python
#Predicted value of y
y_predict = pd.DataFrame(model.predict(X), columns=['y_predict'])
print(y_predict)
```

```python
#Plotting the Hypothesis

plt.figure(figsize=(10, 8))
plt.scatter(X, y, alpha=0.3)
plt.plot(X.to_numpy(), y_predict.to_numpy(), color='red', linewidth=4)
plt.xlim(0, 4.5e8)
plt.ylim(0, 3e9)
plt.xlabel('production_budget_usd')
plt.ylabel('worldwide_gross_usd')
plt.title('Worldwide Gross v/s Production Budget')
plt.show()
```

```python
#R2 Score
r2_score = model.score(X, y)
print(r2_score)
```
